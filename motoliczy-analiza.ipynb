```python
# Cleaning 1 

lata = ['2023']
for rok in lata:

    path = "D:/Dokumenty/Tomek/Data_science/Cepik/nowe_samochody/dane_do_analizy/"+str(rok)+"/"
    out_path = path + "clean_generated_" + datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    files = [file for file in os.listdir(path) if os.path.isfile(os.path.join(path,file))] # Ignore hidden files
    
    
    
    # dataframes
    try:
        del all_data_producenci
        del all_data_modele
        del all_data_marka_bd
        del all_data_elektyczne_producenci
        del all_data_elektyczne_modele
        del all_data_rodzaj_paliwa
    except Exception:
        print("no del " )
    
    all_data_producenci = pd.DataFrame()
    all_data_modele = pd.DataFrame()
    all_data_marka_bd = pd.DataFrame()
    all_data_elektyczne_producenci = pd.DataFrame()
    all_data_elektyczne_modele = pd.DataFrame()
    all_data_rodzaj_paliwa = pd.DataFrame()
    
    try: 
        os.makedirs (out_path)
        print (path_to_save, " successfully created")

    except Exception:
        print("dic: ", out_path, "aready exist" )

        
    try: 

        os.makedirs (out_path + "/modele")
        os.makedirs (out_path + "/Marka_bd")
        os.makedirs (out_path + "/producenci")
        os.makedirs (out_path + "/tylko_nowe")
        os.makedirs (out_path + "/wyczyszczone")
        os.makedirs (out_path + "/elektryczne")
        os.makedirs (out_path + "/rodzaj_paliwa")

    except Exception:
        print("dic: ", out_path, "aready exist" )



    for file in files:
        print ("----------------------")
        print ("file: ", file)
        c = pd.read_csv(path + file, encoding = 'utf-8',sep = ';', 
                        dtype = {'id': object})

        # odrzucam kolumnę
        try:
            cc = c.drop(['Unnamed: 0'], axis = 1)
            print ('odrzucam kolumnę')
        except Exception:
            cc = c
            print("dic: Unname column" )
        

        # usuwam attt. z nagłówków
        cc.columns = cc.columns.str.replace("attributes.","")
        print ("Usuniete 'arrtributes'")

        # nadaję licznik
        cc['liczba_pojazdow']=1
        print ('dodana kolumna licznik  - liczba_pojazdow')

        # odrzucam duplikaty
        print ('-------------')
        print ('odrzucam duplikaty')
        print ('liczba rekordów przed duplikatami', len(cc))
        cc = cc.drop_duplicates()
        print ('liczba rekordów pod odrzuceniu', len(cc))


        # Zamieniam nazwy column repalce names sposob-produkcji na rok-produkcji
        print ('-------------')
        cc.rename(columns = {'sposob-produkcji':'rok-produkcji1'}, inplace = True)
        cc.rename(columns = {'rok-produkcji':'sposob-produkcji'}, inplace = True)
        cc.rename(columns = {'rok-produkcji1':'rok-produkcji'}, inplace = True)
        print ('Zmieniłem nazwy kolumn rok-produkcji - sposob-produkcji')

        # Zamieniam model, na sam model tam gdzie model skłda się z marki i modelu - np. marka: Toyota Yaris na Yaris
        cc['tc_check_model'] = cc.apply(lambda x: str(x['marka']) in str(x['model']), axis=1)
        cc['model'] = cc.apply(lambda x: str(x['model']).replace(str(x['marka']),"") if x['tc_check_model'] == True else x['model'] , axis=1)

        # do przeanalizowania na szybszego np. zamiast lambda.
        #cc['tc_check_model'] = np.where( (cc['marka'] in cc['modek']) ,  zastap , cc['model'])
        #cc['model'] = np.where( (cc['tc_check_model'] == True) ,  cc['marka'] , cc['model'])

        #Zamieniam brak danych w marka na dane z pola nazwa-producenta
        # Zaastępowanie Na w polu Marka nazawą z pola nazwa-producenta

        print ("######### Zamieniam brak danych w marka na dane z pola nazwa-producenta ############### ")
        for dozastapienia, zastap in marki_do_zastapienia_dic.items():

            cc['marka'] = np.where( ((pd.isnull(cc['marka'])) & (cc['nazwa-producenta'] == dozastapienia) ),  zastap , cc['marka'])
            # slooow ! cc['marka'] = cc.apply(lambda x: str(x['marka']).replace(str(x['marka']),zastap) if (pd.isnull(x['marka']) & (x['nazwa-producenta'] == dozastapienia)) else x['marka'] , axis=1) 
            #print ("Zastąpiono: ",dozastapienia, "=> ", zastap)



        print ("Koniec")

        # Zapisuje błedy 
        marka_bd = cc[(pd.isnull(cc['marka'])) & (pd.notnull(cc['nazwa-producenta'])) ]
        all_data_marka_bd = pd.concat([all_data_marka_bd, marka_bd])



        # Usuwam spacje w model i marka
        cc['model'] = cc['model'].str.strip()
        cc['marka'] = cc['marka'].str.strip()

        # Dodaj markę i model - tworząc nowy model
        cc['model'] = cc['marka'] + " " + cc['model']


        # spradzam duplikaty:
        dp = cc.groupby(['id']).count()
        do_zduplikowane = dp[dp['liczba_pojazdow'] > 1]

        #info o duplikatach
        if len(do_zduplikowane) > 1:
            print ("znaleziony duplikaty")
        else :
            print ("NIE znaleziono duplikatów")


        # wypłnij Nulls
        cc['marka'] = cc['marka'].fillna("b.d.")
        cc['model'] = cc['model'].fillna("b.d.")
        cc['rodzaj-paliwa'] = cc['rodzaj-paliwa'].fillna("b.d.")
        cc['rodzaj-pierwszego-paliwa-alternatywnego'] = cc['rodzaj-pierwszego-paliwa-alternatywnego'].fillna("NIE DOTYCZY")


        # ------------- koniec czyszczenia  --------------------

        if zapisac_wyczyszczone == True:
            cc.to_csv(
                path_or_buf = out_path + "/wyczyszczone/Clean_" + file, 
                index = False,
                encoding = 'utf-8',
                )

        # -------------- zapisuję tylko nowe --------------------
        cc_nowe = cc.loc[cc['pochodzenie-pojazdu'].isin(['NOWY IMPORT INDYW', 'NOWY ZAKUPIONY W KRAJU'])]

        # wybieranie wierszy - filtr wierszy pochodzenie pojazdu
        if zapis_tylko_nowe == True:

            cc_nowe.to_csv(
                path_or_buf = out_path + "/tylko_nowe/nowe_" + file, 
                index = False,
                encoding = 'utf-8',
                )


        # -------------- zapisuję tylko NOWE Elektryczne --------------------
        # wybieranie wierszy - filtr wierszy pochodzenie pojazdu
        if zapis_elektryczne == True :
            cc_elektryczne = cc_nowe.loc[cc_nowe['rodzaj-paliwa'].isin(['ENERGIA ELEKTRYCZNA'])]

            cc_elektryczne.to_csv(
                path_or_buf = out_path + "/elektryczne/nowe_ele_" + file, 
                index = False,
                encoding = 'utf-8',
                )



        # ---------------- W Y B Ó R  W I E R S Z Y  -----------------------
        cc_analiza = cc.loc[cc['pochodzenie-pojazdu'].isin(['NOWY IMPORT INDYW', 'NOWY ZAKUPIONY W KRAJU'])] # tylko nowe
        # ---------------- W Y B Ó R  W I E R S Z Y  -----------------------


        # wybór kolumn do analizy
        # wybór kolumny /filter wybiera tylko kolumny - nie filtruje/
        cc_analiza = cc_analiza.filter(items = {'rejestracja-wojewodztwo',
                                                'rodzaj-pojazdu',
                                                'rodzaj-paliwa',
                                                'rodzaj-pierwszego-paliwa-alternatywnego',
                                                'marka',
                                                'model',
                                                'liczba_pojazdow',
                                                'pochodzenie-pojazdu'}, axis =  1)



        cc_analiza.rename(columns =  {'rejestracja-wojewodztwo':'wojewodztwo',
                               'marka' : 'marka',
                               'rodzaj-pojazdu':'rodzaj_pojazdu',
                               'liczba_pojazdow': 'liczba_pojazdow',
                               'rodzaj-paliwa' : 'rodzaj_paliwa',
                               'pochodzenie-pojazdu' : 'pochodzenie_pojazdu',
                                'rodzaj-pierwszego-paliwa-alternatywnego': 'rodzaj_pierwszego_paliwa_alternatywnego'}, 
                              inplace = True)


         # ############## R O DZ A J   P A L I W A  #######################    
        # wybór kolumn do zapisu
        cc_rodzaj_paliwa = cc_analiza.filter(items = {'wojewodztwo',
                                                      'rodzaj_pojazdu',
                                                      'marka', 
                                                      'model' ,
                                                      'rodzaj_paliwa', 
                                                      'rodzaj_pierwszego_paliwa_alternatywnego',
                                                      'liczba_pojazdow'}, axis =  1)

        # grupowanie
        cc_rodzaj_paliwa = cc_rodzaj_paliwa.groupby(['wojewodztwo',
                                                     'rodzaj_pojazdu',
                                                     'marka', 
                                                     'model' ,
                                                     'rodzaj_paliwa', 
                                                     'rodzaj_pierwszego_paliwa_alternatywnego'],as_index=False).sum()
        #kolejność kolumn /istotna dla js/
        #cc_rodzaj_paliwa = cc_rodzaj_paliwa[['wojewodztwo','rodzaj_pojazdu','marka', 'model' ,'rodzaj_paliwa', 'rodzaj_pierwszego_paliwa_alternatywnego','liczba_pojazdow']]

        # build file name 
        out_file_name = file_prefix + "Rodzaj_paliwa_"+str(rok)+"_clean_" + file 

        # Save producenci
        if czy_rodzaj_paliwa == True:
            cc_rodzaj_paliwa.to_csv(
                path_or_buf = out_path + "/rodzaj_paliwa/" +out_file_name, 
                index = False,
                encoding = 'utf-8',
                )
            print ('Zapisano', out_path + out_file_name )
            print ('----------------------')




        # ############## P R O D U C E N C I  #######################    
        # wybór kolumn do zapisu
        cc_producenci = cc_analiza.filter(items = {'wojewodztwo','rodzaj_pojazdu','marka','liczba_pojazdow'}, axis =  1)

        # grupowanie
        cc_producenci = cc_producenci.groupby(['wojewodztwo','marka','rodzaj_pojazdu'],as_index=False).sum()
        #kolejność kolumn /istotana dla js/
        cc_producenci = cc_producenci[['wojewodztwo','rodzaj_pojazdu','marka','liczba_pojazdow']]

        # build file name 
        out_file_name = file_prefix + "Producenci_"+str(rok)+"_clean_" + file 

        # Save producenci
        if zapis_producenci == True:
            cc_producenci.to_csv(
                path_or_buf = out_path + "/producenci/" +out_file_name, 
                index = False,
                encoding = 'utf-8',
                )
            print ('Zapisano', out_path + out_file_name )
            print ('----------------------')

        # ############## M O D E L E  #######################  

        # wybór kolumn do zapisu
        cc_modele = cc_analiza.filter(items = {'wojewodztwo','rodzaj_pojazdu','model','liczba_pojazdow'}, axis =  1)

        # grupowanie
        cc_modele = cc_modele.groupby(['wojewodztwo','model','rodzaj_pojazdu'],as_index=False).sum()

        #Re-order columns
        cc_modele = cc_modele[['wojewodztwo','rodzaj_pojazdu','model','liczba_pojazdow']]
        # build file name 
        out_file_name = file_prefix + "Modele_"+str(rok)+"_clean_" + file

        # SAVE modele 
        if zapis_modele == True:
            cc_modele.to_csv(
                path_or_buf = out_path + "/modele/" +out_file_name, 
                index = False,
                encoding = 'utf-8',
                )
            print ('Zapisano', out_path + out_file_name )
            print ('----------------------')

        # ############## E L K T R Y C Z N E   #######################  

        # wybór kolumn do zapisu
        cc_elektryczne = cc_analiza.loc[cc_analiza['rodzaj_paliwa'].isin(['ENERGIA ELEKTRYCZNA'])] 

        cc_elektryczne_producenci = cc_elektryczne.filter(items = {'wojewodztwo','rodzaj_pojazdu','marka','liczba_pojazdow'}, axis =  1)
        cc_elektryczne_modele = cc_elektryczne.filter(items = {'wojewodztwo','rodzaj_pojazdu','model','liczba_pojazdow'}, axis =  1)


        # grupowanie
        cc_elektryczne_modele = cc_elektryczne_modele.groupby(['wojewodztwo','model','rodzaj_pojazdu'],as_index=False).sum()
        cc_elektryczne_producenci =  cc_elektryczne_producenci.groupby(['wojewodztwo','marka','rodzaj_pojazdu'],as_index=False).sum()

        #Re-order columns
        cc_elektryczne_producenci = cc_elektryczne_producenci[['wojewodztwo','rodzaj_pojazdu','marka','liczba_pojazdow']]
        cc_elektryczne_modele = cc_elektryczne_modele[['wojewodztwo','rodzaj_pojazdu','model','liczba_pojazdow']]

        # SAVE producenci elektryczne

        out_file_name = file_prefix + "Elektryczne_producenci_"+str(rok)+"_clean_" + file
        cc_elektryczne_producenci.to_csv(
            path_or_buf = out_path + "/modele/" +out_file_name, 
            index = False,
            encoding = 'utf-8',
            )
        print ('Zapisano', out_path + out_file_name )
        print ('----------------------')

        # SAVE modele elektryczne
        out_file_name = file_prefix + "Elektryczne_modele_"+str(rok)+"_clean_" + file
        cc_elektryczne_modele.to_csv(
            path_or_buf = out_path + "/modele/" +out_file_name, 
            index = False,
            encoding = 'utf-8',
            )



        # łączenia DF
        all_data_producenci = pd.concat([all_data_producenci, cc_producenci])
        all_data_modele = pd.concat([all_data_modele, cc_modele])
        all_data_elektyczne_producenci = pd.concat([all_data_elektyczne_producenci, cc_elektryczne_producenci])
        all_data_elektyczne_modele = pd.concat([all_data_elektyczne_modele, cc_elektryczne_modele])
        all_data_rodzaj_paliwa = pd.concat([all_data_rodzaj_paliwa, cc_rodzaj_paliwa])


    # Zapis wszystkich danych z WOJEWÓDZTW 
    if zapis_producenci == True:
        all_data_producenci.to_csv(
                path_or_buf = path_agregated_data + "/producenci_"+str(rok)+file_prefix+".csv", 
                index = False,
                encoding = 'utf-8',
                )
    if zapis_modele == True:
        all_data_modele.to_csv(
                path_or_buf = path_agregated_data + "/modele_"+str(rok)+file_prefix+".csv", 
                index = False,
                encoding = 'utf-8',
                )

    all_data_elektyczne_producenci.to_csv(
            path_or_buf = path_agregated_data + "/e_producenci_"+str(rok)+file_prefix+".csv", 
            index = False,
            encoding = 'utf-8',
            )

    all_data_elektyczne_modele.to_csv(
            path_or_buf = path_agregated_data + "/e_modele_"+str(rok)+file_prefix+".csv", 
            index = False,
            encoding = 'utf-8',
            )

    all_data_rodzaj_paliwa.to_csv(
            path_or_buf = path_agregated_data + "/rodzaj_paliwa_"+str(rok)+file_prefix+".csv", 
            index = False,
            encoding = 'utf-8',
            )


    # Zapis jeszcze brakujących nazwa-producenta do macierzy zamiany
    all_data_marka_bd_to_save = all_data_marka_bd[['marka','nazwa-producenta']]
    all_data_marka_bd_to_save = all_data_marka_bd_to_save.drop_duplicates()
    all_data_marka_bd_to_save.to_csv(
            path_or_buf = out_path + "/Marka_bd_"+str(rok)+file_prefix+".csv",
            index = False,
            encoding = 'utf-8',
            )

    print ("Koniec!")
```
